```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

# Load DeepSeek 7B GPTQ model (4-bit quantized to fit in 24GB VRAM)
model_name = "deepseek-ai/deepseek-llm-7b-chat"

# Enable GPU
device = "cuda" if torch.cuda.is_available() else "cpu"

# Load model and tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_name, trust_remote_code=True)
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype=torch.float16,
    device_map="auto"
)

# Move model to GPU
model.to(device)

print("DeepSeek LLM 7B is ready!")
```

```python
prompt = "Generate a recipe using chicken, rice, and broccoli."
inputs = tokenizer(prompt, return_tensors="pt").to(device)

# Generate output
output = model.generate(**inputs, max_length=450)
print(tokenizer.decode(output[0], skip_special_tokens=True))
```

```python
import pandas as pd

from google.colab import drive
drive.mount('/content/drive')

import torch
from datasets import load_dataset
from transformers import AutoModelForCausalLM, AutoTokenizer, TrainingArguments, Trainer
from peft import LoraConfig, get_peft_model
import os

# Enable GPU
device = "cuda" if torch.cuda.is_available() else "cpu"

# Load dataset
file_path = "/content/drive/My Drive/modified_dataset.csv"
dataset = load_dataset("csv", data_files=file_path, split="train")  # Assuming it's pre-tokenized

```

```python
model_name = "deepseek-ai/deepseek-llm-7b-chat"

# Load tokenizer and model
tokenizer = AutoTokenizer.from_pretrained(model_name, trust_remote_code=True)
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype=torch.float16,
    device_map="auto"
)

# Configure LoRA
lora_config = LoraConfig(
    r=8,  # LoRA rank (low-rank matrices)
    lora_alpha=32,
    target_modules=["q_proj", "v_proj"],  # Target layers to fine-tune
    lora_dropout=0.05,
    bias="none",
)

# Apply LoRA to the model
model = get_peft_model(model, lora_config)
model.to(device)

print("LoRA applied to the model. Ready for training!")
```

```python
from datasets import load_dataset

# Load dataset
file_path = "/content/drive/My Drive/modified_dataset.csv"
dataset = load_dataset("csv", data_files=file_path, split="train")

# Preprocess dataset: Combine relevant columns into a single text prompt
def preprocess_function(examples):
    # Create input text in a structured format
    input_text = f"Recipe Title: {examples['title']}\nIngredients: {examples['ingredients']}\nDirections: {examples['directions']}"

    # Tokenize input text
    tokenized_output = tokenizer(input_text, truncation=True, padding="max_length", max_length=512)

    return tokenized_output

# Tokenize dataset
dataset = dataset.map(preprocess_function, remove_columns=dataset.column_names)

# Split dataset: 90% training, 10% validation
dataset = dataset.train_test_split(test_size=0.1)
train_dataset = dataset["train"]
eval_dataset = dataset["test"]

# Update training arguments
training_args = TrainingArguments(
    output_dir="/content/drive/My Drive/deepseek_finetuned",
    per_device_train_batch_size=2,
    gradient_accumulation_steps=4,
    save_total_limit=3,
    num_train_epochs=3,
    logging_dir="./logs",
    logging_steps=50,
    save_strategy="epoch",
    fp16=True,
    warmup_steps=100,
    weight_decay=0.01,
    evaluation_strategy="epoch",
    remove_unused_columns=False,  # ✅ Prevents Trainer from dropping important columns
)

# Define Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=eval_dataset,
)

# Start training
trainer.train()
```

```python
save_path = "/content/drive/My Drive/deepseek_finetuned"
model.save_pretrained(save_path)
tokenizer.save_pretrained(save_path)

print(f"Fine-tuned model saved at {save_path}")
```

```python
prompt = "Generate a recipe using chicken, rice, and broccoli."
inputs = tokenizer(prompt, return_tensors="pt").to(device)
output = model.generate(**inputs, max_length=450)
print(tokenizer.decode(output[0], skip_special_tokens=True))
```
